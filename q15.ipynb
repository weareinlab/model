```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

scene = cv2.imread('church.jpg', cv2.IMREAD_GRAYSCALE)
template = cv2.imread('template.jpg', cv2.IMREAD_GRAYSCALE)

w, h = template.shape[::-1]

result = cv2.matchTemplate(scene, template, cv2.TM_CCOEFF_NORMED)

threshold = 0.8
locations = np.where(result >= threshold)

scene_color = cv2.cvtColor(scene, cv2.COLOR_GRAY2BGR)
for pt in zip(*locations[::-1]):
    cv2.rectangle(scene_color, pt, (pt[0] + w, pt[1] + h), (255, 0, 0), 2)

scene_rgb = cv2.cvtColor(scene_color, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 6))
plt.imshow(scene_rgb)
plt.title("Template Matching Result")
plt.axis('off')
plt.show()
```
