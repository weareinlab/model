```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
model = YOLO('yolov8n.pt')
```

```python
img_path = '/Users/saisrithaja/Desktop/image processing/image.jpg'  
img = cv2.imread(img_path)
```

```python
results = model(img)

annotated_img = results[0].plot() 

plt.imshow(cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB))
plt.axis('off')  
plt.show()

for result in results[0].boxes:
    print(f"Object: {result.cls}, Confidence: {result.conf}, Coordinates: {result.xywh}")
```
